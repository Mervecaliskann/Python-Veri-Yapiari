# Listeler

```
# Boş bir liste oluşturma
bos_liste = []

# Elemanları olan bir liste oluşturma
sayilar = [1, 2, 3, 4, 5]
meyveler = ["elma", "armut", "muz"]
karisik_liste = [1, "merhaba", 3.14, True]

# Listeye eleman ekleme
meyveler.append("çilek")
print(meyveler)  # ['elma', 'armut', 'muz', 'çilek']

# Listeye belirli bir konuma eleman ekleme
sayilar.insert(2, 10)  # 2. indekse 10 ekler
print(sayilar)  # [1, 2, 10, 3, 4, 5]

# Listeden eleman çıkarma (son elemanı)
son_meyve = meyveler.pop()
print(son_meyve)  # çilek
print(meyveler)  # ['elma', 'armut', 'muz']

# Listeden belirli bir konumdaki elemanı çıkarma
silinen_sayi = sayilar.pop(1)  # 1. indeksteki elemanı (2) çıkarır
print(silinen_sayi)  # 2
print(sayilar)  # [1, 10, 3, 4, 5]

# Listeye erişim (indeksleme)
ilk_meyve = meyveler[0]
print(ilk_meyve)  # elma

# Liste dilimleme (slicing)
alt_liste = sayilar[1:4]  # 1. indeksten 4. indekse kadar (4. dahil değil)
print(alt_liste)  # [10, 3, 4]

# Liste uzunluğu
print(len(meyveler))  # 3

# Bir elemanın listede olup olmadığını kontrol etme
if "muz" in meyveler:
    print("Muz listede var.")

# Listeler üzerinde döngü kurma
for meyve in meyveler:
    print(meyve.upper())
```

# Demetler

```
# Demet oluşturma
koordinat = (10, 20)
renkler = ("kırmızı", "yeşil", "mavi")

# Demete erişim (indeksleme)
x = koordinat[0]
y = koordinat[1]
print(x, y)  # 10 20

# Demet dilimleme (slicing)
alt_demet = renkler[0:2]
print(alt_demet)  # ('kırmızı', 'yeşil')

# Demet uzunluğu
print(len(renkler))  # 3

# Demetlerde tek eleman tanımlarken dikkat!
tek_elemanli_demet = ("merhaba",)  # Virgül önemlidir
print(type(tek_elemanli_demet))  # <class 'tuple'>

# Demet paketi açma (tuple unpacking)
isim, yas = ("Ali", 30)
print(isim, yas)  # Ali 30

icice_demet = ((1, 2), (3, 4)) # Demetler başka demetleri de içerebilir
print(icice_demet[0][1]) # İçteki ilk demetin ikinci elemanına erişim: 2

karisik_demet = (1, 2, 2, 3, 2)
tekrar_sayisi = karisik_demet.count(2) # Demette kaç tane '2' olduğunu sayar
print(f"2 sayısı demette {tekrar_sayisi} kez geçiyor")
ilk_indeks = karisik_demet.index(3) # '3' elemanının ilk görüldüğü indeksi bulur
print(f"3 sayısının ilk indeksi: {ilk_indeks}")
```

# Sözlükler

```
# Boş bir sözlük oluşturma
bos_sozluk = {}

# Elemanları olan bir sözlük oluşturma
ogrenci = {
    "ad": "Ayşe",
    "yas": 22,
    "bolum": "Mühendislik"
}

# Sözlüğe erişim (anahtar kullanarak)
ogrenci_adi = ogrenci["ad"]
print(ogrenci_adi)  # Ayşe

# Sözlüğe yeni anahtar-değer çifti ekleme
ogrenci["sehir"] = "İzmir"
print(ogrenci)  # {'ad': 'Ayşe', 'yas': 22, 'bolum': 'Mühendislik', 'sehir': 'İzmir'}

# Sözlükteki bir değerin güncellenmesi
ogrenci["yas"] = 23
print(ogrenci)  # {'ad': 'Ayşe', 'yas': 23, 'bolum': 'Mühendislik', 'sehir': 'İzmir'}

# Sözlükten bir anahtar-değer çiftini silme
del ogrenci["sehir"]
print(ogrenci)  # {'ad': 'Ayşe', 'yas': 23, 'bolum': 'Mühendislik'}

# Bir anahtarın sözlükte olup olmadığını kontrol etme
if "ad" in ogrenci:
    print("Ad anahtarı sözlükte var.")

# Sözlüğün anahtarlarını alma
anahtarlar = ogrenci.keys()
print(anahtarlar)  # dict_keys(['ad', 'yas', 'bolum'])

# Sözlüğün değerlerini alma
degerler = ogrenci.values()
print(degerler)  # dict_values(['Ayşe', 23, 'Mühendislik'])

# Sözlüğün anahtar-değer çiftlerini alma
oge_ciftleri = ogrenci.items()
print(oge_ciftleri)  # dict_items([('ad', 'Ayşe'), ('yas', 23), ('bolum', 'Mühendislik')])

# Sözlük üzerinde döngü kurma (anahtarlar üzerinde)
for anahtar in ogrenci:
    print(anahtar, ":", ogrenci[anahtar])

# Sözlük üzerinde döngü kurma (anahtar-değer çiftleri üzerinde)
for anahtar, deger in ogrenci.items():
    print(f"{anahtar}: {deger}")
```

# Kümeler

```
# Boş bir küme oluşturma
bos_kume = set()

# Elemanları olan bir küme oluşturma
sayilar_kumesi = {1, 2, 3, 4, 5}
meyveler_kumesi = {"elma", "armut", "muz", "elma"}  # "elma" sadece bir kez saklanır
print(meyveler_kumesi)  # {'muz', 'armut', 'elma'} (sıra önemli değil)

# Kümeye eleman ekleme
sayilar_kumesi.add(6)
print(sayilar_kumesi)  # {1, 2, 3, 4, 5, 6}

# Kümeden eleman çıkarma
sayilar_kumesi.remove(3)
print(sayilar_kumesi)  # {1, 2, 4, 5, 6}

# Bir elemanın kümede olup olmadığını kontrol etme
if "muz" in meyveler_kumesi:
    print("Muz kümede var.")

# Kümeler arası işlemler
kume_a = {1, 2, 3, 4}
kume_b = {3, 4, 5, 6}

# Birleşim (union)
birlesim = kume_a.union(kume_b)
print(birlesim)  # {1, 2, 3, 4, 5, 6}

# Kesişim (intersection)
kesisim = kume_a.intersection(kume_b)
print(kesisim)  # {3, 4}

# Fark (difference)
fark_a_b = kume_a.difference(kume_b)
print(fark_a_b)  # {1, 2}

# Simetrik Fark (symmetric difference)
simetrik_fark = kume_a.symmetric_difference(kume_b)
print(simetrik_fark)  # {1, 2, 5, 6}
```

# Dizeler

```
# Dize oluşturma
mesaj = "Merhaba Dünya!"
isim = 'Python'

# Dizeye erişim (indeksleme)
ilk_karakter = mesaj[0]
print(ilk_karakter)  # M

# Dize dilimleme (slicing)
alt_dize = mesaj[0:7]
print(alt_dize)  # Merhaba

# Dize uzunluğu
print(len(mesaj))  # 13

# Dize birleştirme (concatenation)
selamlama = "Hoş geldin, " + isim + "!"
print(selamlama)  # Hoş geldin, Python!

# Dize biçimlendirme (f-strings - Python 3.6+)
yas = 25
bilgi = f"Benim adım {isim} ve {yas} yaşındayım."
print(bilgi)  # Benim adım Python ve 25 yaşındayım.

# Dize metotları (örnekler)
buyuk_harf = mesaj.upper()
print(buyuk_harf)  # MERHABA DÜNYA!

kucuk_harf = mesaj.lower()
print(kucuk_harf)  # merhaba dünya!

ayrilmis_kelimeler = mesaj.split(" ")
print(ayrilmis_kelimeler)  # ['Merhaba', 'Dünya!']

baslangici_kontrol = mesaj.startswith("Mer")
print(baslangici_kontrol)  # True

sonu_kontrol = mesaj.endswith("!")
print(sonu_kontrol)  # True
```
